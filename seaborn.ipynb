```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Ans:- Any five plots that we can  plot using the Seaborn library are:-
1. Scatter Plot : It is a graph that shows the relationship between two variables. It is useful for visualizing patterns
                   or trends in the data and identifying outliers or cluster.           
     
2. Line plot : It shows the changes in a variable over time in graph that is useful for visualizing trends or patterns in
              in time series data.
3. Box plot : it is a  graph that displays the distribution of a variable. It is useful for identifying outliers , visualizing
             and the spread of the data, and comparing the distribution of a variable across different groups.
4. Heatmap : It is a graphical representation of data in which values are represented by colors. It is useful for visualizing
             correlations or patterns in large datasets.
5. Histogram : It is a graph that shows the distribution of a variable by dividing it into bins and plotting the frequency of observations
              in each bin. it is useful for visualizing the shape of the data distribution and identifying outliers.
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
sns.boxplot(x=titanic.pclass , y=titanic.age)
```

```python
sns.boxplot(x='pclass', y ='fare', data=titanic)
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds
```

```python
sns.histplot(x = 'price', data= diamonds, hue = 'cut')
```

```python
sns.scatterplot(x = 'price',y = 'cut', data= diamonds, hue = 'cut')
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights
```

```python
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights, cmap="YlGnBu")
```
